# Example usage
A classic example of MNIST digit classification using `phitodeep`:
## Imports

```python
import numpy as np
from datasets import load_dataset

from phitodeep.layers import base, activation
import phitodeep.loss as loss
import phitodeep.model as m
```

## Load MNIST dataset
Using the `datasets` library from Hugging Face, we can easily load the MNIST dataset:

```python
train_dataset = load_dataset("ylecun/mnist", split="train")
test_dataset = load_dataset("ylecun/mnist", split="test")

X_train = train_dataset["image"]
y_train = train_dataset["label"]
X_test = test_dataset["image"]
y_test = test_dataset["label"]

X_train = np.array(X_train).astype(np.float32) / 255.0
y_train = np.array(y_train)
X_test = np.array(X_test).astype(np.float32) / 255.0
y_test = np.array(y_test)
print(X_train.shape, y_train.shape)
```

## Model Definition
The following two cells show the different ways to initialize the model. Through the `Sequential` class directly, or by using the builder class:

```python
model = m.Sequential(
    base.Flatten(),
    base.Dense(784, 128),
    activation.ReLu(),
    base.Dense(128, 64),
    activation.ReLu(),
    base.Dense(64, 10),
)
model.add(activation.Softmax())
model.setoptimizer("adam")
model.setloss(loss.CategoricalCrossEntropy())
model.setbatchsize(64)

model.summary()
```

```python
model = (
    m.SequentialBuilder()
    .flatten()
    .dense(784, 128)
    .relu()
    .dense(128, 64)
    .relu()
    .dense(64, 10)
    .softmax()
    .optimizer("adam")
    .loss(loss.CategoricalCrossEntropy())
    .alpha(0.05)
    .epochs(5)
    .batch(64)
    .build()
)

model.summary()
```

## Training 
First we check prediction accuracy before training, then we train the model for 5 epochs, and finally check the accuracy again to see the improvement.

```python
y_pred = model.predict(X_test)
accuracy = np.mean(np.argmax(y_pred, axis=1) == y_test) * 100
print(f"Test Accuracy: {accuracy:.4f} %")
```

```python
losses = model.train(X_train, y_train, X_test, y_test)
```

```python
y_pred = model.predict(X_test)
accuracy = np.mean(np.argmax(y_pred, axis=1) == y_test) * 100
print(f"Test Accuracy: {accuracy:.4f} %")
```
